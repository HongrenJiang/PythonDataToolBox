```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import pytz

# 获取Coinbase的分钟级别数据
def get_coinbase_candles(symbol, start_time, end_time, granularity=60):
    url = f'https://api.pro.coinbase.com/products/{symbol}/candles'
    params = {
        'start': start_time.isoformat(),
        'end': end_time.isoformat(),
        'granularity': granularity
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print(f"Error fetching data: {response.status_code} {response.text}")
        return []

# 处理数据
def process_coinbase_data(data):
    df = pd.DataFrame(data, columns=['time', 'low', 'high', 'open', 'close', 'volume'])
    df['time'] = pd.to_datetime(df['time'], unit='s')
    df = df.sort_values('time')
    df.set_index('time', inplace=True)
    return df

# 手动输入时间
def manual_time_input():
    symbol = 'BTC-USD'  # 可以替换为其他符号，如'ETH-USD', 'SOL-USD'

    # 手动输入开始和结束时间，格式为 'YYYY-MM-DD HH:MM:SS'
    start_time_input = '2023-05-01 15:00:00'
    end_time_input = '2023-05-01 17:00:00'

    # 将输入的时间字符串转换为 datetime 对象，并设置为 EST 时区
    est = pytz.timezone('US/Eastern')
    start_time = est.localize(datetime.strptime(start_time_input, '%Y-%m-%d %H:%M:%S')).astimezone(pytz.utc)
    end_time = est.localize(datetime.strptime(end_time_input, '%Y-%m-%d %H:%M:%S')).astimezone(pytz.utc)

    # 获取数据
    data = get_coinbase_candles(symbol, start_time, end_time)

    # 处理数据
    if data:
        df = process_coinbase_data(data)
        print(df)
        return df
    else:
        print("No data fetched")
        return None

# 调用函数
df = manual_time_input()

# 如果需要可视化，可以添加绘图代码
if df is not None and not df.empty:
    plt.figure(figsize=(12, 6))
    plt.bar(df.index, df['volume'], width=0.0005)
    plt.xlabel('Time (EST)')
    plt.ylabel('Volume')
    plt.title('Trading Volume from 3 PM to 5 PM (EST) on 2023-05-01')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 获取并处理数据
def get_and_process_data(symbol, start_time, end_time):
    data = get_coinbase_candles(symbol, start_time, end_time)
    if data:
        df = process_coinbase_data(data)
        return df
    else:
        print(f"No data fetched for {symbol}")
        return pd.DataFrame(columns=['time', 'low', 'high', 'open', 'close', 'volume'])


def manual_time_input_2():
    symbols = {
        "BTC-USD": "Bitcoin",
        "ETH-USD": "Ethereum",
        "SOL-USD": "Solana"
    }

    # 手动输入开始和结束时间，格式为 'YYYY-MM-DD HH:MM:SS'
    start_time_input = '2023-05-01 15:00:00'
    end_time_input = '2023-05-01 17:00:00'

    # 将输入的时间字符串转换为 datetime 对象，并设置为 EST 时区
    est = pytz.timezone('US/Eastern')
    start_time = est.localize(datetime.strptime(start_time_input, '%Y-%m-%d %H:%M:%S')).astimezone(pytz.utc)
    end_time = est.localize(datetime.strptime(end_time_input, '%Y-%m-%d %H:%M:%S')).astimezone(pytz.utc)

    data = {}
    for symbol, name in symbols.items():
        df = get_and_process_data(symbol, start_time, end_time)
        if not df.empty:
            # 计算这两小时的总交易量
            total_volume = df['volume'].sum()
            # 将每分钟的交易量转换为百分比
            df['volume_percent'] = df['volume'] / total_volume * 100
            data[name] = df

    # 绘制Histogram
    plt.figure(figsize=(14, 10))
    for name, df in data.items():
        if not df.empty:
            plt.bar(df.index, df['volume_percent'], alpha=0.5, label=name, width=0.0005)

    plt.xlabel('Time (EST)')
    plt.ylabel('Volume Percentage')
    plt.title('Trading Volume Percentage from 3 PM to 5 PM (EST) on 2023-05-01')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

# 调用新的函数
manual_time_input_2()
```

```python
# 新的函数manual_time_input_3()
def manual_time_input_3():
    symbols = {
        "BTC-USD": "Bitcoin",
        "ETH-USD": "Ethereum",
        "SOL-USD": "Solana"
    }

    # 手动输入开始和结束时间，格式为 'YYYY-MM-DD HH:MM:SS'
    start_time_input = '2023-05-01 15:00:00'
    end_time_input = '2023-05-01 17:00:00'

    # 将输入的时间字符串转换为 datetime 对象，并设置为 EST 时区
    est = pytz.timezone('US/Eastern')
    start_time = est.localize(datetime.strptime(start_time_input, '%Y-%m-%d %H:%M:%S')).astimezone(pytz.utc)
    end_time = est.localize(datetime.strptime(end_time_input, '%Y-%m-%d %H:%M:%S')).astimezone(pytz.utc)

    data = {}
    for symbol, name in symbols.items():
        df = get_and_process_data(symbol, start_time, end_time)
        if not df.empty:
            # 计算这两小时的总交易量
            total_volume = df['volume'].sum()
            # 将每分钟的交易量转换为百分比
            df['volume_percent'] = df['volume'] / total_volume * 100
            data[name] = df

    # 创建子图
    fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(14, 18), sharex=True)
    
    for ax, (name, df) in zip(axes, data.items()):
        if not df.empty:
            ax.bar(df.index, df['volume_percent'], alpha=0.5, label=name, width=0.0005)
            ax.set_ylabel('Volume Percentage')
            ax.set_title(f'{name} Trading Volume Percentage')
            ax.legend()
    
    plt.xlabel('Time (EST)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# 调用新的函数
manual_time_input_3()
```
